```python
# Importing the libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
dataset = pd.read_csv("/Users/balogunishaq/Desktop/Complete Machine Learning Course - Siddhardhan/datasets/breast-cancer-dataset.csv")
```

```python
dataset
```

```python
dataset.head()
```

```python
# Finding the count of different labels
dataset['Breast'].value_counts()
```

```python
# Load the Label Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(dataset.Breast)
```

```python
# Append the labels to the DataFrame
dataset['target'] = labels
```
